```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car_data = pd.read_csv('car_sales_data.csv')
```

```python
dropped_rows_data = car_data.dropna(axis=0) 
print(dropped_rows_data)
```

```python
unique_data = dropped_rows_data.drop_duplicates()
print(unique_data)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(car_data['Price ($)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Car Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(car_data['Annual Income'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Annual Incomes')
plt.xlabel('Annual Income ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
gender_counts = car_data['Gender'].value_counts()
plt.figure(figsize=(8, 5))
gender_counts.plot(kind='bar', color=['lightcoral', 'lightskyblue'])
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
model_counts = car_data['Model'].value_counts().head(10)  # Taking the top 10 models for clarity
plt.figure(figsize=(10, 6))
model_counts.plot(kind='bar', color='orange')
plt.title('Top 10 Car Models')
plt.xlabel('Car Model')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
transmission_counts = car_data['Transmission'].value_counts()
plt.figure(figsize=(8, 5))
transmission_counts.plot(kind='bar', color='lightseagreen')
plt.title('Distribution of Transmission Types')
plt.xlabel('Transmission Type')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
color_counts = car_data['Color'].value_counts().head(10)  # Taking the top 10 colors for clarity
plt.figure(figsize=(10, 6))
color_counts.plot(kind='bar', color='lightcoral')
plt.title('Top 3 Car Colors')
plt.xlabel('Color')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Distribution of dealer regions
dealer_region_counts = car_data['Dealer_Region'].value_counts()
plt.figure(figsize=(12, 6))
dealer_region_counts.plot(kind='bar', color='lightblue')
plt.title('Distribution of Dealer Regions')
plt.xlabel('Dealer Region')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Distribution of companies
company_counts = car_data['Company'].value_counts().head(10)  # Taking the top 10 companies for clarity
plt.figure(figsize=(10, 6))
company_counts.plot(kind='bar', color='orange')
plt.title('Top 10 Car Companies')
plt.xlabel('Company')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Distribution of body styles
body_style_counts = car_data['Body Style'].value_counts()
plt.figure(figsize=(10, 6))
body_style_counts.plot(kind='bar', color='lightcoral')
plt.title('Distribution of Body Styles')
plt.xlabel('Body Style')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=car_data,y='Company',palette = 'pastel')
```
